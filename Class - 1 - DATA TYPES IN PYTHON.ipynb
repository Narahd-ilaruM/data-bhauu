## Let's talk about numbers!

* A lot of types of numbers are supported in Python like integers (int type) and decimal numbers (float type).


* Integers are just whole numbers, positive or negative. For example: 41 and -70


* Float numbers are those that have a decimal point in them, or use an exponential (E) to define the number. For example 3.0 and -22.1 are examples of floating point numbers. 4E7 (4 times 10 to the power of 7) is also an example of a floating point number in Python.

##  Rules for naming a variable in Python

* Variables names must start with a letter or an underscore like _product , product_


* The remainder of your variable name may consist of letters, numbers and underscores


* shrey1, HarSH,Deep_learning are some valid variable names


* Python is case sensitive.


* case_sensitive, CASE_SENSITIVE, and Case_Sensitive are each a different variable.

* Names cannot begin with a number.


* Names can not contain spaces.


* Names can not contain any of these symbols:

      :'",<>/?|\!@#%^&*~-+


* Avoid using Python built-in keywords like `list`, `str` , `def` etc.
